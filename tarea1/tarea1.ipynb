```python
'''
Se eligió el dataset: Spotify Tracks Dataset (CSV)

Ya que contiene bastante información de miles de canciones disponibles
en Spotify, incluyendo: Nombre de la canción, Artista, Año de lanzamiento,
Popularidad, Duración, etc.

La razón por la que tomé este dataset es debido a que quiero retomar el análisis
de modelos predictivos en el ámbito musical.

En particular, con este conjunto se contienen tanto variables numéricas como
categóricas, lo que permite aplicar filtros, estadísticas descriptivas
y operaciones aritméticas. Además de que posee un tamaño bastante considerable.
'''
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Spotify Tracks Analysis") \
    .getOrCreate()

spotify_df = spark.read.csv(
    "/content/dataset.csv",
    header=True,
    inferSchema=True
)

spotify_df.printSchema()
spotify_df.show(5)
```

```python
'''
Operaciones
'''
```

```python
# Canciones populares
popular_tracks = spotify_df.filter(spotify_df.popularity > 70)
popular_tracks.select("track_name", "artists", "popularity").show(5)

```

```python
# Estadísticas sobre variables num
spotify_df.describe(
    "popularity",
    "duration_ms",
    "danceability",
    "energy"
).show()
```

```python
# Conteo de canciones por géenero
spotify_df.groupBy("track_genre") \
    .count() \
    .orderBy("count", ascending=False) \
    .show(10)
```
