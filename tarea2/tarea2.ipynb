```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Spotify RDD Analysis") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
rdd = sc.textFile("/content/dataset.csv")
```

```python
rdd.take(5)
```

```python
# Preprocesamiento del RDD
header = rdd.first()
data_rdd = rdd.filter(lambda row: row != header)

rows = data_rdd.map(lambda row: row.split(","))
```

```python
# Conteo total de canciones
total_tracks = rows.count()
total_tracks
```

```python
# Canciones por genero
genre_counts = rows.map(lambda r: (r[-1], 1)) \
                   .reduceByKey(lambda a, b: a + b)

genre_counts.take(10)

```

```python
#Conversión RDD a Dataframe
from pyspark.sql import Row

genre_df = genre_counts.map(lambda x: Row(genre=x[0], count=x[1])).toDF()
genre_df.show(10)
```
